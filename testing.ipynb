**Perbandingan Harga Saham Aktual dan Prediksi**

```python
import matplotlib.pyplot as plt
import numpy as np

# Contoh data
dates = np.arange(1, 11)  # Misalnya 10 hari terakhir
actual_prices = [100, 102, 105, 107, 108, 110, 112, 114, 116, 118]
predicted_prices = [99, 101, 104, 106, 109, 110, 113, 115, 117, 119]

# Membuat plot
plt.figure(figsize=(10,6))
plt.plot(dates, actual_prices, label='Harga Aktual', color='blue', marker='o')
plt.plot(dates, predicted_prices, label='Harga Prediksi', color='red', linestyle='--', marker='x')

# Menambahkan label dan judul
plt.title('Perbandingan Harga Saham Aktual dan Prediksi')
plt.xlabel('Tanggal')
plt.ylabel('Harga Saham')
plt.legend()
plt.grid(True)
plt.show()
```

Visualisasi Metrik Evaluasi

```python
# Contoh data metrik
labels = ['BBCA (80/20)', 'BBCA (70/30)', 'BBCA (60/40)',
          'BBRI (80/20)', 'BBRI (70/30)', 'BBRI (60/40)',
          'BBNI (80/20)', 'BBNI (70/30)', 'BBNI (60/40)']
mse_values = [0.0055, 0.0005, 0.0035, 0.0010, 0.0007, 0.0017, 
              0.0010, 0.0006, 0.0048]
rmse_values = [0.074, 0.0226, 0.0591, 0.0319, 0.0262, 0.0416, 
               0.0319, 0.0255, 0.0694]
mae_values = [0.0698, 0.0178, 0.0523, 0.0261, 0.0203, 0.0354, 
              0.0258, 0.0192, 0.0613]

# Membuat subplots
fig, ax = plt.subplots(figsize=(12, 8))

# Membuat bar chart
width = 0.2  # lebar bar
x = np.arange(len(labels))  # posisi x

bar1 = ax.bar(x - width, mse_values, width, label='MSE', color='b')
bar2 = ax.bar(x, rmse_values, width, label='RMSE', color='r')
bar3 = ax.bar(x + width, mae_values, width, label='MAE', color='g')

# Menambahkan label dan judul
ax.set_xlabel('Model')
ax.set_ylabel('Nilai Metrik')
ax.set_title('Perbandingan Metrik Evaluasi: MSE, RMSE, dan MAE')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```

Hasil Prediksi vs Aktual - Visualisasi Tren Masa Depan

```python
# Data contoh untuk prediksi masa depan
future_dates = np.arange(11, 16)  # 6 bulan berikutnya
future_predicted_prices = [120, 122, 125, 127, 129, 131]

# Membuat plot
plt.figure(figsize=(10,6))
plt.plot(dates, actual_prices, label='Harga Aktual', color='blue', marker='o')
plt.plot(dates, predicted_prices, label='Harga Prediksi', color='red', linestyle='--', marker='x')
plt.plot(future_dates, future_predicted_prices, label='Prediksi Masa Depan', color='green', linestyle=':', marker='s')

# Menambahkan label dan judul
plt.title('Perbandingan Harga Saham Aktual, Prediksi, dan Prediksi Masa Depan')
plt.xlabel('Tanggal')
plt.ylabel('Harga Saham')
plt.legend()
plt.grid(True)
plt.show()
```
